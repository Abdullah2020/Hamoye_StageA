#Question 12

#Question 13
What is the total number and percentage of missing data in 2016 to 2 decimal places?


```python


missing_data_Y2016 = food_df.isnull().sum().loc['Y2016']
percent_missing = round((missing_data_Y2016 * 100 / len(food_df)), 2)

print('The total number of missing data in 2016 = ', missing_data_Y2016)
print('The % of the missing data in 2016 = ', percent_missing,'%')
```
